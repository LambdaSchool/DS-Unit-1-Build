```
import pandas as pd
import plotly.offline as py
import plotly.graph_objects as go

plus_one = pd.read_csv('Prem_Cost_Employee_plus_One.csv', skiprows=2, skipfooter=14, engine='python')
family = pd.read_csv('Prem_Cost_Family.csv', skiprows=2, skipfooter=12, engine='python')
single = pd.read_csv('Prem_Cost_Single_Employee.csv', skiprows=2, skipfooter=12, engine='python')
```

```
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```
family_1 = family.drop(labels=[0])
family_1['code'] = states
family_1.head()
```

```
fig1 = go.Figure(data=go.Choropleth(
    locations=family_1['code'],
    z = family_1['2018__Employer Contribution'].astype(float),
    locationmode = 'USA-states',
    colorscale = 'Blues'
))

fig1.update_layout(
    title_text = 'Average Premium Cost Paid by Employers For Family Plans',
    geo_scope='usa'
)

fig1.show()
```

```
py.offline.plot(fig1, filename='employer_family.html')
```
