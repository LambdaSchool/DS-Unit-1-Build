```
# I need to do this in an asyncronous event loop for some reason

# This will grab data from Bungie's API and construct a dataframe
# containing all of the information on base weapon stats.
# Data pulled on 3/1/2020
async def main():
  destiny = pydest.Pydest('0b1855e0fd624d25aac8fbabc0f51421')

  # resulting list of data that'll be turned into a dataframe
  resultslist = []

  for weap in wp:
    #print('grabbing hash:',weap)

    dt = await destiny.decode_hash(weap, 'DestinyInventoryItemDefinition')

    # Set up variables to grab data
    row = {}

    # Grab the name of the item first
    row['Name'] = dt['displayProperties']['name']
    #print('name:', row['name'])

    for i, entries in enumerate(dt['stats']['stats'].values()):
      # Stats of weapons
      # statHash is the name of the stat
      # value is the base value
      temp = (await destiny.decode_hash(entries['statHash'], 'DestinyStatDefinition'))
      if temp['displayProperties']['name'] == '':
        row['unnamed_stat'] = entries['value']
      else:
        row[temp['displayProperties']['name']] = entries['value']
      #print(entries)

    # Get the weapon type
    row['Weapon Type'] = dt['itemTypeDisplayName']

    row['Rarity'] = dt['itemTypeAndTierDisplayName'][:-(len(row['Weapon Type']))].strip()

    if 'sockets' in dt: # catch bad weapons
      for i, entries in enumerate(dt['sockets']['socketEntries']):
        # Perks on weapons
        # singleInitialItemHash is the curated roll perk for that slot
        if entries['singleInitialItemHash'] == 0:
          row['Perk ' + str(i) + " Init"] = np.NaN
        else:
          temp = (await destiny.decode_hash(entries['singleInitialItemHash'], 'DestinyInventoryItemDefinition'))
          row['Perk ' + str(i) + " Init"] = temp['displayProperties']['name']

        # randomizedPlugSetHash is the list of randomly rolled perks for that slot
        if 'randomizedPlugSetHash' in entries:
          row['Perk ' + str(i) + ' Rolls Hash'] = entries['randomizedPlugSetHash']

        # reusablePlugSetHash indicates that the slot should always use the curated roll 
        # (i.e. intrinsics, exotic weapons)
        if 'reusablePlugSetHash' in entries:
          row['Perk ' + str(i) + ' is Intrinsic'] = True
        else:
          row['Perk ' + str(i) + ' is Intrinsic'] = False

        # turn roll hashes into perk lists
        if 'Perk ' + str(i) + ' Rolls Hash' in row:
          templist = await destiny.decode_hash(row['Perk ' + str(i) + ' Rolls Hash'], 'DestinyPlugSetDefinition')
          templist = templist['reusablePlugItems']
          #print(templist)
          hashlist = []
          for j, item in enumerate(templist):
            hashlist.append(await destiny.decode_hash(templist[j]['plugItemHash'],'DestinyInventoryItemDefinition'))
            hashlist[j] = hashlist[j]['displayProperties']['name']
          #print(hashlist)
          row['Perk ' + str(i) + ' Rolls'] = hashlist
    else: # fill bad weapons with NaNs
      for i in range(10):
        row['Perk ' + str(i) + ' Init'] = np.NaN
        row['Perk ' + str(i) + ' Rolls Hash'] = np.NaN
        row['Perk ' + str(i) + ' is Intrinsic'] = np.NaN

    # Append it to the results list
    resultslist.append(row)

  resultslist = pd.DataFrame(resultslist, index=wp)

  #print(row)

  # close the loop
  await destiny.close()

  return resultslist

# Set up and run the event loop
loop = asyncio.get_event_loop()
df = loop.run_until_complete(main()) # snag the data
loop.close()
```
